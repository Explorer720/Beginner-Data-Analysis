```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1=pd.read_csv('data1.csv')
```

```python
x=data1['University']
y=data1['Total']
plt.plot(x,y)
plt.title("New Students Enrollment in Different Universities")
plt.xlabel('Universities')
plt.ylabel('New Students Enrollment')
plt.show()
```

```python
x=data1['University']
y=data1['Total']
plt.bar(x,y)
plt.title("New Students Enrollment in Different Universities")
plt.xlabel('Universities')
plt.ylabel('New Students Enrollment')
plt.show()
```
